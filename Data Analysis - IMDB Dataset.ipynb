## 22. Find Unique values from Genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]                   # 2d list to 1d list 
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
        
```

```python
uni_list
```
